## Get dictionary from csv list
In the dataset there are 88 csv file. Firstly I categorized the csv files into 4 category named `Participant Data` , `Race Time Data`, `Session Data` , `Telemetry Data` then get one dictionary for each cateogry using the following steps:
*         All non integer values are replaced by 'ZZZ'
*         All numerical values below (equal) the mean value are replaced by 'BA'
*         All numerical values greater the mean value are replaced by 'AA'
*         Outliers (data whose distance from mean is three times greater than standard deviation) are replaced by 'O'
